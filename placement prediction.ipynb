```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
df = pd.read_csv("C:\\Users\\vipul\\Downloads\\collegePlace.csv")
```

```python
df
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
```

```python
# Assuming 'df' is your DataFrame and 'stream' is the column containing stream information
Stream_mapping = {
    'Electronics And Communication': 0,
    'Computer Science': 1,
    'Mechanical': 2,
    'Civil': 3,
    'Information Technology': 4,
    'Electrical': 5
}

df['Stream'] = df['Stream'].map(Stream_mapping)
```

```python
df
```

```python
# Split the data into features (X) and target variable (y)
X = df.drop('PlacedOrNot', axis=1)
y = df['PlacedOrNot']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Model Selection: Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model using metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print("Model Evaluation:")
print("Accuracy:", accuracy*100)
print("Precision:", precision*100)
print("Recall:", recall*100)
print("F1-score:", f1*100)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Model Training
def train_model(df):
    # Split the data into features (X) and target variable (y)
    X = df[['Age', 'Gender', 'Stream', 'Internships', 'CGPA', 'Hostel', 'HistoryOfBacklogs']]
    y = df['PlacedOrNot']
    
    # Model Selection: Logistic Regression
    model = LogisticRegression()
    model.fit(X, y)
    
    return model

# Function to predict placement based on user input
def predict_placement(model, user_input):
    # Create a DataFrame with user input
    data = pd.DataFrame(user_input, index=[0])
    
    # Make prediction
    prediction = model.predict(data)
    
    return prediction[0]

# Example usage:
def main():
    
    
    # Train the model
    model = train_model(df)
    
    # Get user input
    user_input = {}
    user_input['Age'] = int(input("Enter Age: "))
    user_input['Gender'] = int(input("Enter Gender (1 for Male, 0 for Female): "))
    user_input['Stream'] = int(input("Enter Stream (0 for Electronics And Communication, 1 for Computer Science, etc.): "))
    user_input['Internships'] = int(input("Enter Number of Internships: "))
    user_input['CGPA'] = float(input("Enter CGPA: "))
    user_input['Hostel'] = int(input("Enter Hostel (1 for Yes, 0 for No): "))
    user_input['HistoryOfBacklogs'] = int(input("Enter History of Backlogs: "))
    
    # Predict placement
    placement_prediction = predict_placement(model, user_input)
    
    print("Predicted Placement (1 for Yes, 0 for No):", placement_prediction)

if __name__ == "__main__":
    main()
```
